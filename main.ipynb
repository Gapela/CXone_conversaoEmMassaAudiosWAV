```python
import subprocess
import os
```

```python
# functions
def convert_audio(input_audio, output_audio):
    try:
        command = [
            'ffmpeg',
            '-i', input_audio,
            '-ar', '8000',           # Sample rate
            '-ac', '1',              # Mono
            '-codec:a', 'pcm_mulaw', # μ-law codec
            output_audio
        ]

        result = subprocess.run(command, check=True, capture_output=True, text=True)
        print(result.stdout)

    except subprocess.CalledProcessError as e:
        print("Erro ao converter o áudio:")
        print(e.stderr)
    except Exception as e:
        print("Erro inesperado:", e)

def paths_creation(path_input, path_output):
    try:
        if not os.path.isdir(path_input):
            os.makedirs(path_input, exist_ok=True)
        if not os.path.isdir(path_output):
            os.makedirs(path_output, exist_ok=True)
    except Exception as e:
        print("Erro ao criar caminhos de input e output: ", e)
    
def path_listing(dir):
    try:
        if not os.path.isdir(dir):
            raise FileNotFoundError(f"Diretório não encontrado: {dir}")
        else:
            list_audios = sorted([f for f in os.listdir(dir) if f.lower().endswith('.wav')])
            return list_audios
    except Exception as e:
        print("Erro ao listar diretorio: ", e)

def convert_all_audios(list_audios, path_input, path_output):
    try:
        print("list_audios: ", list_audios)
        print("path_input: ", path_input)
        print("path_output: ", path_output)

        if not list_audios:
            print("Nenhum arquivo .wav encontrado.")
            return

        for audio in list_audios:
            print("Iniciando conversão do audio:", audio)
            try:
                convert_audio(os.path.join(path_input, audio), os.path.join(path_output, audio))
                print("     system: sucesso")
            except Exception as e:
                print(f"     erro ao converter '{audio}':", e)

        print("Conversão finalizada!")
    
    except Exception as e:
        print("Erro em converter todos os audios:", e)
```

```python
# paths
path_input = "" # insira o caminho com os audios a serem convertidos NOME DA PASTA DEVE SER "input"
path_output = "" # insira o caminho para os audios convertidos serem salvos NOME DA PASTA DEVE SER "output"
```

```python
# LISTING DIR
paths_creation(path_input, path_output) # checa se a pasta input e output existe, se não exisir vai criar
list_audios = path_listing(path_input) # faz lista dos audios dentro do path input
convert_all_audios(list_audios, path_input, path_output) # converte todos os audios listados na pasta input e salva na pasta output
```
